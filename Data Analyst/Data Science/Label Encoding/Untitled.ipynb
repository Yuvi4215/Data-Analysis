# Lable Encoding

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
le=LabelEncoder()
```

```python
data=pd.read_csv("../../DataSets/cars93.csv")
```

```python
data.head()
```

```python
data["Type"].unique()
```

## label encoding ordinal data(by alphabetical order)

```python
data["Type"]=le.fit_transform(data["Type"])
```

```python
data.head()
```

## label encoding ordinal data(custom order)

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
data=pd.read_csv("../../DataSets/cars93.csv")
```

```python
ord=OrdinalEncoder(categories=[[ 'Compact','Small', 'Midsize','Sporty', 'Van', 'Large' ]])
data["Type"]=ord.fit_transform(data[["Type"]]).astype("int")
```
