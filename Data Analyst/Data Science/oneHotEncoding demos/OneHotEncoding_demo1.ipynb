```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
data=pd.DataFrame({
    "Hero":["T","W","B","I","S","L"],
    "Movie":["Marvel","Disney","Marvel","Disney","Marvel","Paramount"],
    "Seuqel":[4,2,8,3,2,6]
})
```

```python
data.head()
```

```python
ohe=OneHotEncoder()
result=ohe.fit_transform(data[["Hero"]]).toarray()
result=pd.DataFrame(result,columns=["B","I","L","S","T","W"])
data1=pd.concat([data,result],axis=1)
data1.drop("Hero",axis=1,inplace=True)
data1
```

```python
data1=pd.get_dummies(data,columns=["Hero"],dtype="int",drop_first=True)
data1
```
