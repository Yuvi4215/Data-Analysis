## 1. Import Libraries & Global Configuration

```python

import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = (8,5)
plt.rcParams['axes.titlesize'] = 14
plt.rcParams['axes.labelsize'] = 12
```

## 2. Basic Line Plot

```python

x = np.arange(1, 11)
y = x ** 2

plt.plot(x, y)
plt.title("Basic Line Plot")
plt.show()
```

## 3. Styled Line Plot

```python

plt.plot(x, y, linestyle='--', marker='o', linewidth=2)
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Styled Line Plot")
plt.grid(True)
plt.show()
```

## 4. Multiple Lines in One Plot

```python

y2 = x ** 3

plt.plot(x, y, label="xÂ²")
plt.plot(x, y2, label="xÂ³")
plt.legend()
plt.title("Multiple Line Plot")
plt.show()
```

## 5. Scatter Plot

```python

x_sc = np.random.rand(50)
y_sc = np.random.rand(50)

plt.scatter(x_sc, y_sc)
plt.title("Scatter Plot")
plt.show()
```

## 6. Bar Chart

```python

categories = ['A','B','C','D']
values = [25, 40, 30, 35]

plt.bar(categories, values)
plt.title("Bar Chart")
plt.show()
```

## 7. Bar Chart with Value Labels

```python

bars = plt.bar(categories, values)

for bar in bars:
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height(),
             bar.get_height(),
             ha='center',
             va='bottom')

plt.title("Bar Chart with Labels")
plt.show()
```

## 8. Horizontal Bar Chart

```python

plt.barh(categories, values)
plt.title("Horizontal Bar Chart")
plt.show()
```

## 9. Histogram

```python

data = np.random.normal(50, 10, 1000)

plt.hist(data, bins=20, edgecolor='black')
plt.title("Histogram")
plt.show()
```

## 10. Histogram with Density

```python

plt.hist(data, bins=20, density=True)
plt.title("Histogram with Density")
plt.show()
```

## 11. Pie Chart

```python

labels = ['Python','Java','C++','JavaScript']
sizes = [40,30,15,15]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title("Pie Chart")
plt.show()
```

## 12. Subplots

```python

fig, ax = plt.subplots(1, 2, figsize=(12,4))

ax[0].plot(x, y)
ax[0].set_title("Line Plot")

ax[1].bar(categories, values)
ax[1].set_title("Bar Plot")

plt.tight_layout()
plt.show()
```

## 13. Grid Customization

```python

plt.plot(x, y)
plt.grid(True, linestyle=':', alpha=0.7)
plt.title("Grid Customization")
plt.show()
```

## 14. Axis Limits

```python

plt.plot(x, y)
plt.xlim(0, 12)
plt.ylim(0, 120)
plt.title("Axis Limits")
plt.show()
```

## 15. Log Scale

```python

plt.plot(x, y2)
plt.yscale('log')
plt.title("Log Scale Plot")
plt.show()
```

## 16. Annotations

```python

plt.plot(x, y)
plt.annotate("Maximum Value",
             xy=(10, 100),
             xytext=(6, 60),
             arrowprops=dict(arrowstyle="->"))
plt.title("Annotations")
plt.show()
```

## 17. Saving Figures

```python

plt.plot(x, y)
plt.title("Saving Figure Example")
plt.savefig("example_plot.png", dpi=300, bbox_inches='tight')
plt.show()
```

## 18. Using Style Sheets

```python

plt.style.use('ggplot')

plt.plot(x, y)
plt.title("ggplot Style")
plt.show()

plt.style.use('default')
```

## 19. Error Bar Plot

```python

errors = np.random.rand(len(x)) * 5

plt.errorbar(x, y, yerr=errors, fmt='o')
plt.title("Error Bar Plot")
plt.show()
```

## 20. Step Plot

```python

plt.step(x, y)
plt.title("Step Plot")
plt.show()
```

## 21. Stack Plot

```python

y1 = np.array([20,35,30,35,27])
y2 = np.array([25,32,34,20,25])
x2 = np.arange(len(y1))

plt.stackplot(x2, y1, y2)
plt.title("Stack Plot")
plt.show()
```

## 22. Fill Between

```python

plt.plot(x2, y1)
plt.fill_between(x2, y1, alpha=0.3)
plt.title("Fill Between")
plt.show()
```

## 23. Twin Axis Plot

```python

fig, ax1 = plt.subplots()

ax1.plot(x2, y1)
ax1.set_ylabel("Y1")

ax2 = ax1.twinx()
ax2.plot(x2, y2)
ax2.set_ylabel("Y2")

plt.title("Twin Axis Plot")
plt.show()
```

## 24. Object-Oriented (OO) API

```python

fig, ax = plt.subplots()

ax.plot(x, y)
ax.set_title("Object Oriented API")
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.grid(True)

plt.show()
```

## 25. Clear and Close Figures

```python

plt.plot(x, y)
plt.show()

plt.clf()   # Clear figure
plt.close() # Close window
```
