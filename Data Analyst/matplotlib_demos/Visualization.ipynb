```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr1=np.arange(1,11)
arr2=arr1** 2
```

### Line Graph

```python
plt.plot(arr1, arr2)
plt.xlabel("Numbers")
plt.ylabel("Squre")
plt.title("Number vs Squre")
#plt.savefig("demo.jpg")
plt.figure(figsize=(10,10))
```

```python
plt.plot(arr1, arr2, marker="x",markerfacecolor="red", markeredgecolor="red", color="black")
```

### hist graph

```python
plt.hist(data1["Price"],edgecolor="white",color="black", bins=12)
plt.title("Price vs Count")
plt.xlabel("Price")
plt.ylabel("Count")
```

```python
data1["Type"].value_counts()
```

```python
x= data1["Type"].value_counts().index
y=data1["Type"].value_counts().values
```

```python
plt.bar(x,y,edgecolor="white",color="black")
plt.title("Type vs Count")
plt.xlabel("Car Type")
plt.ylabel("Count")
```

```python
data1.groupby("Type")["Price"].mean()
```

```python
data1.groupby("AirBags")["Price"].mean()
```

```python
values=data1.groupby("AirBags")["Price"].mean().values
label=data1.groupby("AirBags")["Price"].mean().index
```
