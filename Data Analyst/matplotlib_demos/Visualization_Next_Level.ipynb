```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
arr1=np.arange(1,26)
arr2=arr1** 2
```

```python
plt.figure(figsize=(10, 6))  # Create figure first

plt.plot(arr1, arr2, marker='o', linestyle='--', linewidth=2)

plt.xlabel("Numbers", fontsize=12)
plt.ylabel("Square", fontsize=12)
plt.title("Number vs Square", fontsize=14, fontweight='bold')

plt.grid(True, linestyle=':', alpha=0.7)
plt.tight_layout()

# plt.savefig("demo.jpg", dpi=300)
plt.show()
```

```python
plt.figure(figsize=(10, 6))

plt.plot(
    arr1, arr2,
    marker='o',
    linestyle='--',
    linewidth=2,
    markersize=6
)

plt.xlabel("Numbers", fontsize=12, labelpad=8)
plt.ylabel("Square", fontsize=12, labelpad=8)
plt.title("Relationship Between Number and Its Square",
          fontsize=15, fontweight='bold', pad=10)

plt.grid(visible=True, which='major',
         linestyle=':', linewidth=1, alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
# Create figure
fig, ax = plt.subplots(figsize=(10, 6))

# Plot data
ax.plot(
    arr1, arr2,
    marker='o',
    linewidth=2.5,
    markersize=6
)

# Labels & title
ax.set_xlabel("Number", fontsize=12)
ax.set_ylabel("Square of Number", fontsize=12)
ax.set_title(
    "Relationship Between a Number and Its Square",
    fontsize=15,
    fontweight="bold",
    pad=12
)

# Grid & aesthetics
ax.grid(True, linestyle="--", alpha=0.6)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Layout
fig.tight_layout()

# Save (optional)
# fig.savefig("number_vs_square.png", dpi=300)

plt.show()
```

```python
data= pd.read_csv("../DataSets/Iris.csv")
```

```python
plt.figure(figsize=(8, 6))  # Always first

plt.scatter(
    data["Petal_Length"],
    data["Petal_Width"],
    s=60,          # marker size
    alpha=0.7,     # transparency
    edgecolors='k'
)

plt.title("Petal Length vs Width", fontsize=14, fontweight='bold')
plt.xlabel("Petal Length", fontsize=12)
plt.ylabel("Petal Width", fontsize=12)

plt.grid(True, linestyle=':', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(9, 6))

scatter = plt.scatter(
    data["Petal_Length"],
    data["Petal_Width"],
    c=data["Petal_Width"],     # color mapped to values
    cmap="viridis",
    s=80,
    alpha=0.8,
    edgecolors="black"
)

plt.title("Relationship Between Petal Length and Width",
          fontsize=15, fontweight="bold")


plt.xlabel("Petal Length (cm)", fontsize=12)
plt.ylabel("Petal Width (cm)", fontsize=12)

plt.colorbar(scatter, label="Petal Width Intensity")

plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

```python
data1= pd.read_csv("../DataSets/cars93.csv")
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

ax.hist(
    data1["Price"],
    bins=12,
    edgecolor="white",
    linewidth=1.2
)

ax.set_title(
    "Distribution of Prices",
    fontsize=15,
    fontweight="bold",
    pad=12
)
ax.set_xlabel("Price", fontsize=12)
ax.set_ylabel("Count", fontsize=12)

ax.grid(axis="y", linestyle="--", alpha=0.6)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

fig.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

counts, bins, patches = ax.hist(
    data1["Price"],
    bins=20,
    color="grey",
    edgecolor="white",
    linewidth=1.2
)

# Add count labels on top of bars
for count, patch in zip(counts, patches):
    ax.text(
        patch.get_x() + patch.get_width() / 2,
        patch.get_height(),
        int(count),
        ha="center",
        va="bottom",
        fontsize=10
    )

ax.set_title(
    "Distribution of Prices",
    fontsize=15,
    fontweight="bold",
    pad=12
)
ax.set_xlabel("Price", fontsize=12)
ax.set_ylabel("Count", fontsize=12)

ax.grid(axis="y", linestyle="--", alpha=0.6)
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

fig.tight_layout()
plt.show()
```
